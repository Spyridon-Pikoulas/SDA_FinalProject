## Obesity Rates in New York

So, obesity rates in New York City. But before we do that, it's important to understand what obesity and obesity rates actually mean. Obesity is a medical condition where a person has an excessive amount of body fat, which can lead to various health problems. Obesity rates refer to the percentage of people in a specific area who are classified as obese.

To help visualize the obesity rates across New York City's five boroughs, we'll be using a color-coded map, commonly known as a choropleth map. This map will display a range of colors from light red to deep red, each representing the obesity rate in a particular borough. The deeper the red color, the higher the obesity rate in that area, while lighter red or yellowish colors indicate lower obesity rates.

Apparently, Manhattan has the lowest obesity rate among New York City's five boroughs at 17%. In contrast, the Bronx has the highest rate at 33%, and Staten Island follows closely with a rate of 31%. Brooklyn and Queens fall in between, with obesity rates of 25% and 24%, respectively. Notably, the Bronx and Staten Island have almost double the obesity rates of Manhattan, thus some other factors may contribute to these differences. Let's explore them!
