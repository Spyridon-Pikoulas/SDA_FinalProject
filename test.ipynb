```python
nob_df = ntas.merge(ob_df, left_on="boro_name", right_on="Name", how='left')
```

```python
bor_df = nob_df.dissolve(by="boro_name", aggfunc="first")[["geometry", "Name", "Obesity"]]
bor_df
```

# Choropleth Map

```python
bor_geo = bor_df.to_json()
m = folium.Map(location=[40.7128, -74.0060], zoom_start=11, control_scale=True)
```

```python
folium.Choropleth(
    geo_data=bor_geo,
    data=bor_df,
    columns=["Name", "Obesity"],
    key_on="feature.properties.Name",
    fill_color="YlOrRd",
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name='Obesity Rates by Borough in New York City',
).add_to(m)
```

```python
# Add the names of thr Boroughs
from folium.features import DivIcon

for idx, row in bor_df.iterrows():
    folium.map.Marker(
        location=[row.geometry.centroid.y, row.geometry.centroid.x],
        icon=DivIcon(
            icon_size=(150, 36),
            icon_anchor=(7, 20),
            html=f"<div style='font-size: 12pt; color: black;'>{row.name}</div>",
        ),
    ).add_to(m)
```
